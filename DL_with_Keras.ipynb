# Let's create dummy datasets and
# run the data through Numpy, Scikit learn

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

```python
#Creating dummy dataset for 5% of younger group that experienced side effects
for i in range(50):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)   

#Creating dummy dataset for 5% of older group that did not experience side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
    
#Creating dummy dataset for 95% of younger group that did not experience side effects
for i in range(1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)

#Creating dummy dataset for 95% of older group that did experience side effects    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)     
```

```python
for i in train_samples:
    print(i)
```

```python
# Put into numpy arrays and shuffle the data
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

# Next we'll work with the TensorFlow/Keras to get
# our neural network sequential model ready

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
#building the sequential model, which has a linear stack of layers
model = Sequential([
    Dense(units=16, input_shape=(1,), activation='relu'),  #this is first hidden layer that also creates input layer
    Dense(units=32, activation='relu'),                #this is second hidden layer
    Dense(units=2, activation='softmax')               #this is output layer, did patient have side effects or not?
])
```

```python
model.summary()
```

# Now we'll train the model on the random data we created

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=scaled_train_samples, y=train_labels, batch_size=10, epochs=30, shuffle=True, verbose=2)
```

```python
# Within 30 epochs we are within 93% accuracy rate
# You want low loss and high accuracy so you can generalize on other, new data.
```

```python
# Epochs = the number times that the learning algorithm will work through the entire training dataset. 
# One epoch means that each sample 
# in the training dataset has had an opportunity to update the internal model parameters.
```

```python
# But we need to validate our model on different data.
# We don't want overfitting of our model here, which means our model
# works well on the trained data but not generalize- give good results- on validation data.
```

```python
# refit our model to split out portion of training data into a validation data
# make sure your training data is shuffled before you use validation_split parameter
model.fit(x=scaled_train_samples, y=train_labels, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=2)
```

```python
#Our model is not overfitting and is generalizing well because the 
#training accuracy is about the same as the value accuracy
```

# Making a confusion matrix to see how well we can
# make predictions during inference on new data.

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true=test_labels, y_pred=rounded_predictions)
```

```python
#directly copied from sklearn's website

def plot_confusion_matrix(cm, classes,
                        normalize=False,
                        title='Confusion matrix',
                        cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no_side_effects', 'had_side_effects']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
```

```python
#the top left and bottom right diagonals are the correct predictions
```
